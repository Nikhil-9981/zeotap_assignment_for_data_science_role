```python
import pandas as pd


```

```python
# Load the uploaded datasets
customers_path = "Customers.csv"
products_path = "Products.csv"
transactions_path = "Transactions.csv"

customers = pd.read_csv(customers_path)
products = pd.read_csv(products_path)
transactions = pd.read_csv(transactions_path)


customers_info = customers.info()
products_info = products.info()
transactions_info = transactions.info()

customers_info, products_info, transactions_info
```

```python
# Convert date columns to datetime format for easier manipulation
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

customers['SignupDate']
```

```python
# Basic EDA: Descriptive statistics and checking for missing values
customers_summary = customers.describe(include='all')
products_summary = products.describe(include='all')
transactions_summary = transactions.describe(include='all')
customers_summary
```

```python
products_summary
```

```python
customers_missing = customers.isnull().sum()
products_missing = products.isnull().sum()
transactions_missing = transactions.isnull().sum()

```

```python
transactions_missing
```

```python
products_missing
```

```python
customers_missing
```

```python
# Joining datasets for a holistic analysis
transactions_with_details = transactions.merge(customers, on='CustomerID', how='left').merge(products, on='ProductID', how='left')

transactions_with_details
```

```python
# Aggregating key metrics
total_revenue = transactions_with_details['TotalValue'].sum()
total_customers = customers['CustomerID'].nunique()
total_products_sold = transactions_with_details['Quantity'].sum()
revenue_by_region = transactions_with_details.groupby('Region')['TotalValue'].sum()
top_selling_products = transactions_with_details.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(5)
top_revenue_products = transactions_with_details.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(5)


```
